```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df.columns
```

```python
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]
y=df['koi_disposition']
X=df.drop(['koi_disposition'],axis=1)
X.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
X_train.head()
```

# Scaling X and y

```python
# Scale your data
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
X_scaler=MinMaxScaler().fit(X_train)
X_train_sc=X_scaler.transform(X_train)
X_test_sc=X_scaler.transform(X_test)

y_encoder=LabelEncoder().fit(y_train)
y_train_encod=y_encoder.transform(y_train)
y_test_encod=y_encoder.transform(y_test)

from keras.utils import to_categorical
y_train_cat=to_categorical(y_train_encod)
y_test_cat=to_categorical(y_test_encod)
```

# Deep Keras Model -the Best Model


```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
number_inputs=40
number_output=3
nodes_number=100
model.add(Dense(units=nodes_number, activation='relu',input_dim=number_inputs))
model.add(Dense(units=nodes_number, activation='relu'))
model.add(Dense(units=number_output, activation='softmax'))
```

# Train this Model

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train_sc,y_train_cat, epochs=20,shuffle=True, verbose=2)
```

```python
model_loss, model_accuracy=model.evaluate(X_train_sc,y_train_cat, verbose=2)
print(model_loss)
print(model_accuracy)
```

# Predictions

```python
new_data=X_test_sc[:20]
new_data_y=y_test[:20]
pr=model.predict_classes(new_data)
res_coded=y_encoder.inverse_transform(pr)
result=pd.DataFrame({"actual":new_data_y,"predicted":res_coded})
result
```

```python
#Save the Model
import joblib
filename = 'deep_model.sav'
joblib.dump(model, filename)
```

# Support Vector Machine Model

```python
from sklearn.svm import SVC
modelSVC=SVC(kernel='linear')
modelSVC.fit(X_train_sc,y_train_encod)
modelSVC.score(X_train_sc,y_train_encod)
```

```python
pred=modelSVC.predict(X_test_sc)
from sklearn.metrics import classification_report
print (classification_report(y_test_encod,pred, target_names=['CANDIDATE','CONFIRMED','FALSE POSITIVE']))
```

```python
pred_cat=y_encoder.inverse_transform(pred)
table=pd.DataFrame({"Actual":y_test[:20],"Predicted":pred_cat[:20]})
table
```

```python
#Save the Model
joblib.dump(modelSVC,"SVC_model.sav")
```

```python
print(f"Training Data Score: {modelSVC.score(X_train_sc, y_train_encod)}")
print(f"Testing Data Score: {modelSVC.score(X_test_sc, y_test_encod)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,3,100], 'gamma':[0.0001, 0.0005, 0.001, 0.005]}
grid=GridSearchCV(modelSVC, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_sc,y_train_encod)
#grid.estimator.get_params().keys()
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
pr11=grid.predict(X_test_sc)
pr11_cat=y_encoder.inverse_transform(pr11)
SVC_grig_df=pd.DataFrame({"Actual":y_test[:20],"Predicted":pr11_cat[:20]})
SVC_grig_df
```

# KNN Model


```python
from sklearn.neighbors import KNeighborsClassifier
test_scores=[]
train_scores=[]
for k in range(10,80,5):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_sc,y_train_encod)
    train_score=knn.score(X_train_sc,y_train_encod)
    train_scores.append(train_score)
    test_score=knn.score(X_test_sc,y_test_encod)
    test_scores.append(test_score)
    print(f"k:{k} train_score: {train_score} test_score: {test_score}")
```

```python
import matplotlib.pyplot as plt
plt.plot(range(10,80,5), train_scores, marker='*', color='red')
plt.plot(range(10,80,5), test_scores, marker='x', color='blue')
plt.show()
```

```python
# k=20 => after that situation became more or less stable for a short time
modelKNN=KNeighborsClassifier(n_neighbors=25)
modelKNN.fit(X_train_sc,y_train_encod)
modelKNN.score(X_train_sc,y_train_encod)
#TakeAway- weak model
```

```python
pr2=modelKNN.predict(X_test_sc)
pr2_cat=y_encoder.inverse_transform(pr2)
KNNdf=pd.DataFrame({"Actual":y_test[:20],"Predicted":pr2_cat[:20]})
KNNdf
```
